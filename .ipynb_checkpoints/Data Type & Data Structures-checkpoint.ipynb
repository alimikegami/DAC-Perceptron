Method or function for string objects:
object.upper() -> string menjadi huruf kapital semua, tidak mengoverwrite string aslinya, jika ingin permanen maka assign                         kembali
object.lower() -> menjadi huruf kecil untuk seluruh karakter
object.split() -> memisahkan string menjadi list berdasarkan whitespace (default) atau berdasarkan karakter yang                         diberikan pada parameternya, ex object.split('i') <- karakter pada parameter akan diremove



## String Formatting with .format()

```python
print('Hello {}, this is {}'.format('World', 'Alim'))
```

```python
print('Hello {1}, this is{0}'.format('World', 'Alim'))
```

```python
print('Hello {w}, this is {a}'.format(w = 'World', a = 'Alim'))
```

## Float Formatting

Float formatting: "{value:width.precision f}"
Ex: {r:0.3f}

```python
res = 7/3
```

```python
res
```

```python
print("The result is {r:0.3f}".format(r = res))
```

## F String Literals

```python
name = 'Alim Ikegami'
age = 20
print(f'{name} is {age} years old')
```
