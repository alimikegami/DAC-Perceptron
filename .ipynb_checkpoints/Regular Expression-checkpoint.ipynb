```python
import re
```

```python
# regex to search particular pattern, ex: phone number, email format
phone_num = '0361-999-128'
res = re.search(r'\d\d\d\d-\d\d\d-\d\d\d', phone_num)
```

```python
res.string
```

```python
# Shorten version
res_two = re.search(r'\d{4}-\d{3}-\d{3}', phone_num)
```

```python
res_two.string
```

```python
print(res_two.group())
```

```python
res_two.start() #starting index of matching pattern
```

## Compile method

compile method allows us to access certain pattern only while combining the whole pattern to search through the iterable object

```python
phone_pattern = re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
phone_num = '036-999-1281'
res = re.search(phone_pattern, phone_num)
```

```python
res.group()
```

```python
res.group(1)
```

```python
res.group(2)
```

## Additional Regex Syntax

```python
# OR operator using pipe operator
re.search(r'cat|dog', 'the dog is here') # search for cat OR dog
```

```python
# Wildcard character
re.findall(r'.at', 'the cat inside the hat went splat') #Xat
```

```python
re.findall(r'...at', 'the cat inside the hat went splat') #Xat
```

```python
# search some string that starts with a number
re.findall(r'^\d', '1 is my favorite number')
```

```python
# ends with a number
re.findall(r'\d$', 'the number is 3')
```

```python
# exclude digits or numbers
phrase = 'my favorite number is 34 or 17'
```

```python
pattern = r'[^\d]'
```

```python
res = re.findall(pattern, phrase)
```

```python
print(res)
```

```python
pattern = r'[^\d]+'
res = re.findall(pattern, phrase)
print(res)
```

```python
test_str = 'this is a test phrase to remove the punctuation! and whitespace, and commas'
```

```python
res = re.findall(r'[^ !,]+', test_str)
```

```python
res
```

```python
' '.join(res)
```

```python
text = 'only find the hypen-words in this sentence. But you do not know how long-ish they are'
```

```python
for match in re.findall(r'\w*-\w*', text):
    print(match)
```
